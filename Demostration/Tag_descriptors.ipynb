```python
import numpy as np
from matplotlib import pyplot as plt
from  scipy.signal import convolve2d
def sinusoid(x,f):
    return np.cos(2*np.pi*f*x)

f1 = 1/float(2*np.pi)
f2 = 4*f1
period1 = 1/f1
period2 = 1/f2
print period1
print period2
x = np.linspace(0,period1,int(np.floor(2*period1)))
sinf1 = sinusoid(x,f1)
sinf2 = sinusoid(x,f2)
plt.plot(sinf1)
plt.plot(sinf2)
plt.show()

discf1 = np.sign(sinf1)
discf2 = np.sign(sinf2)
plt.plot(discf1)
plt.plot(discf2)
plt.show()

tag = np.outer(discf1,discf2)

def sampler(image,n=10):
    x,y = image.shape
    samplesX = []
    samplesY = []
    for i in range(n):
        samplesX +=  [ image[0,int(i*(y/float(n)))] ]
        samplesY +=  [ image[int(i*(x/float(n))),0] ]
    return samplesX, samplesY

def sample_value(samples):
    acc = 0
    n = len(samples)
    prev = samples[0]
    for i in range(1,n):
        if prev-samples[i]!=0:
            acc+=1
        prev = samples[i]
    return acc
        
        
        
plt.imshow(tag)
plt.show()
samplesX, samplesY = sampler(tag,n=int(np.floor(2*period1)))
xval,yval = sample_value(samplesX),sample_value(samplesY)
print xval,yval
plt.plot(samplesX)
plt.plot(samplesY)
plt.show()




```
